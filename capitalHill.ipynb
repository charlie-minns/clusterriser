```python
import numpy as np
import matplotlib.pyplot as plt
import hdbscan
```

```python
data = np.load("Data/subset.npy")
data = np.moveaxis(data, 0, -1).reshape(-1, 191)
n = data.shape[1]

x, y = np.meshgrid(0.1*np.arange(200), 0.1*np.arange(200))
arrays = [x.flatten(), y.flatten()]
for i in range(n):
    arrays.append(data[:, i])
data = np.stack(arrays, axis=1)
```

```python
clusterer = hdbscan.HDBSCAN().fit(data)
labels = clusterer.labels_.reshape(200, 200)
plt.imshow(labels)
```

```python
clusterer = hdbscan.HDBSCAN(cluster_selection_method='leaf').fit(data)
labels = clusterer.labels_.reshape(200, 200)
plt.imshow(labels)
```
